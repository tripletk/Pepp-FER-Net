```python
from keras.models import load_model
from keras.preprocessing import image

import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
```

```python
model = load_model('model-with-data-augmentation.hdf5')
```

```python
test_img_gs = cv2.imread('face_min.jpg',0)
print(test_img_gs)
```

```python
plt.imshow(test_img_gs)
plt.show()
```

```python
plt.imshow(test_img_gs, cmap='gray')
plt.show()
```

```python
test_images = []
image = np.array(test_img_gs, dtype=int)
image = np.reshape(image, (96,96,1))
test_images.append(image)
test_images = np.array(test_images)/255.
pred = model.predict(test_images)
```

```python
def plot_sample(image, keypoint, axis, title):
    image = image.reshape(96,96)
    axis.imshow(image, cmap='gray')
    axis.scatter(keypoint[0::2], keypoint[1::2], marker='x', s=20)
    plt.title(title)
```

```python
fig = plt.figure(figsize=(20,16))
axis = fig.add_subplot(4, 5, 0+1, xticks=[], yticks=[])
plot_sample(test_images[0], pred[0], axis, "")
plt.show()
```
